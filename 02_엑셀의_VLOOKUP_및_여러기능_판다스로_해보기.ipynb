## 1. 엑셀 파일에서 데이터프레임 불러오기(read_excel)

> pandas read_excel(io, sheet_name=0, header=0, index_col=None)

엑셀파일을 데이터프레임으로 불러오는 함수

---
**io**

파일의 경로명을 받아주는 인자가 io. pd.read_excel(url)에서 변수 url이 io 인자에 입력된 인수

---
**sheet_name** (인수는 문자열, 정수, 리스트 / 기본값은 0)

엑셀에서 불러올 시트를 지정하는 인자


예)

* 지정하지 않을 때: 첫번째 시트를 불러온다

* 1: 2번째 시트를 불러온다. sheet_name=1 이렇게.

* "Sheet1": 문자열을 입력하면 해당 이름을 가진 시트를 불러온다. 여기서는 "Sheet1"이라는 이름의 시트를 불러온다

* [0, 1, "Sheet5"]: 첫번째 시트와 두번째 시트 그리고 "Sheet5"라는 이름의 시트 세개를 딕셔너리로 통합해 가져온다.

* None: 모든 시트를 딕셔너리로 통합해 가져온다.

---
**header** (인수는 정수, 정수의 리스트 / 기본값은 0)

columns를 지정하는 인자. 지정하지 않으면 맨 윗줄이 columns가 된다. 리스트로 지정하면 멀티 인덱스인 columns가 된다.

---
**index_col** (인수는 정수, 정수의 리스트 / 기본값은 None)

index를 지정하는 인자. 지정하지 않으면 RangeIndex가 index로 부여된다. 리스트로 지정하면 멀티 인덱스인 index가 된다.

## 2. 단일요건 vlookup (merge)

**vlookup 이란?**

<img src=https://i.ibb.co/ZYnKKps/vlookup.png, width=600>

* 두 데이터프레임을 기준 열의 내용에 따라 병합하는 엑셀 함수

* 엑셀에서 가장 유용한 함수로 꼽힌다

* df.merge(df2, how='left', on= '음료') 일 때

 df1에 merge를 하는데, 병합할 기준 열을 on 뒤에 적고, how 뒤에 병합 방식을 적은 후, df2를 df1에 merge한다.



> padas merge

# df.merge(right, how='inner', on=None)

**right** (인수 데이터프레임 혹은 시리즈)

병합할 객체

**how** (인수는 ‘left’, ‘right’, ‘outer’, ‘inner’, ‘cross’ / 기본값은 ‘inner’)

병합할 방식을 결정하는 인자.

* left: 왼쪽 데이터프레임의 키(key)만을 병합에 사용한다.(엑셀의 vlookup과 유사)

* right: 오른쪽 데이터프레임의 키(key)만을 병합에 사용한다.

* outer: 양쪽 데이터프레임의 키(key)들의 합집합을 병합에 사용한다.

* inner: 양쪽 데이터프레임의 키(key)들의 교집합을 병합에 사용한다.

* cross: 양쪽 데이터프레임의 곱집합(cartesian product)을 생성한다.


**on** (인수는 열의 레이블 또는 리스트 / 기본값은 None)

병합의 기준이 되는 열을 지정하는 인자. 기본 값은 양쪽 데이터프레임에서 이름이 공통인 열들이 지정된다. 리스트로 입력하면 복수의 열을 기준으로 병합한다.

## 3. 다중요건 vlookup (merge)

**다중 요건 vlookup 이란?**

<img src=https://i.ibb.co/MP73fmX/vlookup.png, width=600>

* 복수의 열을 기준으로 vlookup을 하는 것

* 위 그림의 경우 업체와 음료 모두 기준 열로 삼아 두 데이터프레임을 병합하였다. 가격을 업체와 음료 두 개의 조건을 붙여야 한다.

* df1.merge(df2, how='left', on=['업체', '음료'])

<img src=https://i.ibb.co/4W4KYvF/image.jpg, width=600>

```python
import pandas as pd
pd.options.display.max_rows = 6 # 판다스 버전업에 따라 6행만 출력의 바뀐 코드. 출력은 6줄까지만 하겠단 의미
data1 = [['송중기', '별다방', '아이스티'],
         ['김나현', '콩다방', '카페오레'],
         ['권보아', '별다방', '카페오레'],
         ['박효신', '콩다방', '아이스티'],
         ['김범수', '별다방', '카페오레']]
data2 = [['별다방', '아이스티', 4500],
         ['별다방', '카페오레', 4000],
         ['콩다방', '아이스티', 6000],
         ['콩다방', '카페오레', 5500]]
df1 = pd.DataFrame(data1, columns=['이름', '업체', '음료'])
df2 = pd.DataFrame(data2, columns=['업체', '음료', '가격'])
```

```python
# df1과 df2를 업체와 음료 두개의 열로 vlookup 하기
# 가격이 뒤에 붙은 새로운 DataFrame이 만들어진다.

df1.merge(df2, how='left', on=['업체', '음료'])
```

```python
# 기본 값으로 양쪽 데이터프레임에서 공통인 열이 모두 on으로 지정된다
# on은 기본적으로 공통인 열 이름은 모두 default로 지정을 해주기에
# on=['업체', '음료']이 없어도 결과는 똑같이 나온다.

df1.merge(df2, how='left')
```

```python
# 엑셀파일로 한번 더 실습해보자
url = '/content/02_03_merge02.xlsx'
df3 = pd.read_excel(url)
df4 = pd.read_excel(url, sheet_name=1) #url 변수에 담긴 excel 파일의 두 번째 시트

# 이번엔 업체마다 동일 제품의 가격이 다를 때는
#'제품'만 on에 넣어서는 안된다. '업체'와 '제품' 모두 on에 넣어 다중요건 vlookup을 사용해야 한다.
# 하지만 문제 발생
# 문제1 : 다중요건 vlookup을 하려고 하는데 각 표에서 해당 요건의 순서가 다르다.
#   한쪽은 제품, 업체 순이고, 다른 쪽은 업체, 제품 순이다. -> 이건 엑셀에선 문제가 되는데, pandas에선 문제 없다.
# 문제2 : '가격'만 붙이는 게 아니라 '구분'도 같이 붙여주고 싶다.
    # 즉, 두 개의 열을 동시에 vlookup하고 싶다 -> 이것도 pandas에선 문제 없다.
```

```python
# df3와 df4를 다중요건 vlookup을 한 뒤 엑셀파일로 저장하자
# df3.merge(df4, how='left', on = ['업체', '제품']) 인데,['업체', '제품']은 df3와 df4의 공통열이기 때문에 없어도 같은 결과가 나온다.

# 아래 데이터 프레임을 엑셀로 저장하고 싶다면 .to_excel('result_merge02.xlsx', index=False) 붙여라
# 파일 이름은 'result_merge02.xlsx' 라는 의미고, index는 가져가지 않겠다는 의미로 index=False

#GPT야 인덱스를 가져가지 않는다는 게 무슨 의미인지 알려줘!

df3.merge(df4, how='left').to_excel('result_merge02.xlsx', index=False)
```

## 4. 행과 열의 이름 바꾸기(rename)

> pandas rename


**index** (인수는 mapper)

index의 이름을 바꾸는 인자

* mapper란? 딕셔너리나 시리즈나 함수와 같이 맵핑을 할수 있는 매개체.
 특정 키 값을 넣으면 반드시 하나의 값이 나와야 한다.

---


**columns** (인수는 mapper)

columns의 이름을 바꾸는 인자


---


**level** (인수는 level의 로케이션 혹은 레이블 /  기본값은 None)

멀티인덱스에서 이름을 바꿀 레벨을 지정하는 인자. 기본값은 모든 level에서 이름을 바꾼다.


---


[rename 함수 설명 블로그](https://kimpanda.tistory.com/72)

[rename 판다스 공식문서](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rename.html)

<img src=https://i.ibb.co/mFXNJwg/image.jpg, width=600>

```python
import pandas as pd
pd.options.display.max_rows = 6 # 판다스 버전업에 따라 6행만 출력의 바뀐 코드
data1 = [['송중기', '별다방', '아이스티'],
         ['김나현', '콩다방', '카페오레'],
         ['권보아', '별다방', '카페오레'],
         ['박효신', '콩다방', '아이스티'],
         ['김범수', '별다방', '카페오레']]
data2 = [['별다방', '아이스티', 4500],
         ['별다방', '카페오레', 4000],
         ['콩다방', '아이스티', 6000],
         ['콩다방', '카페오레', 5500]]
df1 = pd.DataFrame(data1, columns=['이름', '업체', '음료'])
df2 = pd.DataFrame(data2, columns=['업체', '제품', '가격'])

#문제점. df1에서는 '음료' 열에 음료들의 데이터가 있는데,
# df2에서는 '제품' 열에 음료들의 데이터가 담겨 있다.
#on을 쓸 때는 양쪽 데이터프레임에서 동일한 이름의 열을 기준으로 삼았는데, 열의 이름이 다를 경우 문제가 된다.
# 이럴 경우를 대비해 좌측과 우측의 데이터 프레임 기준 열을 merge 함수 인자로 지정할 수 있다.
```

```python
# 제품 열을 음료 열로 열의 이름을 바꿔보자. 인자 안에는 매퍼를 넣어야 하는데 매퍼는 일반적으로 딕셔너리 구조다.
df2.rename(columns={'제품':'음료'})
```

```python
# 내가 놓친 개념!! df2를 실행해보라(함수의 결과는 원본을 덮어쓰지 않는다)
# 판다스 함수의 경우 df에 rename을 적용한다고 해도 함수 적용 결과가 생기는 것이지 기존 데이터 프레임은 바뀌는 게 아니다.
df2
```

<img src=https://i.ibb.co/hWXRMcJ/inplace.png, width=400>

```python
# 내가 놓친 개념!! 함수의 적용결과가 원본을 덮어쓰지 않기에 새로운 변수로 지정해야 한다
# 판다스의 원본 덮어쓰기(inplace) 개념 중요하다
# 판다스에서는 대부분의 함수를 적용한 결과가 원본을 덮어쓰지 않는다. (극히 일부 예외 제외)
# 함수 적용 결과를 기존 변수에 다시 저장하거나 inplace=True를 사용해야 원본을 덮어쓰게 된다.

df3 = df2.rename(columns={'제품':'음료'})
df1.merge(df3, how='left')
```

```python
# 혹은 함수를 적용 결과를 그대로 사용해도 된다.
# 강사는 덮어쓰는 과정을 추가하지 않아서이게 더 편리한 방법이라고 한다.
# 판다스는 빅데이터를 다루기 때문에 덮어쓸 때마다 데이터가 영속적으로 변형된다.
df1.merge(df2.rename(columns={'제품':'음료'}), how='left')
```

---

정렬 순서를 오름차순으로 정렬할지 내림차순으로 정렬할지 지정하는 인자

## 6. 범위로 병합하기 (merge_asof).
강사왈: 엑셀에선 vlookup true라는 정확한 값이 아닌 근사값으로 vlookup을 해주는 작업을 판다스에서도 할 수 있거든요.

 판다스에서는 vlookup은 merge로 하지만 vlookup true 근사값 vlookup은 merge_asof라는 별도의 함수가 있습니다.

 #GPT야  강사의 설명을 정리해서 정확하고 이해하기 쉬운 설명으로 다듬어줘!




> pandas merge_asof


pd.merge_asof(left, right, on=None, by=None, direction='backward')

---

함수(function)와 메서드(method)


사실 판다스에서는 모든 함수는 pandas.으로 시작한다.


pandas를 pd로 쓰기로 했기 때문에 pd.으로 시작한다 (import pandas as pd)

다만 몇몇 함수는 메서드로 사용할 수 있다.


pd.merge(df1, df2)      → 함수로 사용 가능  


df1.merge(df2)          → 메서드도 가능

pd.merge_asof(df1, df2) → 함수로만 가능


📌 일반적으로 여러 함수를 이어서 쓸 때 메서드가 사용하기 편리합니다!


---

**left** (인수는 데이터프레임 혹은 시리즈)

병합할 객체1

---

**right** (인수는 데이터프레임 혹은 시리즈)

병합할 객체2

---

**on** (인수는 열의 레이블)

유사일치로 병합할 기준이 되는 열의 레이블을 지정한다.

* 반드시 하나의 열만 지정해야 한다.
* 숫자나 datetime 같이 유사일치가 가능한 자료형의 열이어야 한다.
* 오름차순으로 정렬이 되어 있어야 한다.

---

**by** (인수는 열의 레이블 또는 열의 레이블의 리스트)

정확히 일치시킬 열을 지정하는 인자.

---

**direction** (인수는 ‘backward’, ‘forward’, ‘nearest’ / 기본값은 ‘backward’)

경계를 기준으로 유사일치시킬 방향을 결정하는 인자


* ‘backward’는 작은 값중 최대를 찾아 병합한다. 기본값이다.

* ‘forward’는 큰 값중 최소를 찾아 병합한다.

* ‘nearest’는 가장 가까운 값을 찾아 병합한다.


---

```python
import pandas as pd
pd.options.display.max_rows = 6 # 판다스 버전업에 따라 6행만 출력의 바뀐 코드
data1 = [['가', 88], ['나', 72], ['다', 80], ['라', 60], ['마', 95]]
data2 = [[90, 'A'], [80, 'B'], [70, 'C'], [0, 'F']]
df1 = pd.DataFrame(data1, columns=['이름', '점수'])
df2 = pd.DataFrame(data2, columns=['점수', '학점'])
df2
```

```python
# 강사왈: merge_asof를 쓰려면
# 점수를 기준으로 범위로 병합한다 (반드시 오름차순으로 정렬되어 있어야함)
df1 = df1.sort_values('점수')
df2 = df2.sort_values('점수')
pd.merge_asof(df1, df2, on='점수')
```

```python
# 엑셀파일로 merge_asof 실습하기
url = '/content/02_06_merge_asof.xlsx'
df3 = pd.read_excel(url)  #엑셀 첫 번째 시트
df4 = pd.read_excel(url, sheet_name=1).sort_values('수량')   #엑셀 두 번째 시트.
# 전체 범위로 정렬해줘야 함.
```

```python
# 사이즈와 종류는 정확히 일치시키고 수량은 범위로 병합한다
# 강사: by는 정확하게 일치시키겠다는 데이터를 집어넣으면 된다.

# # GPT 추가설명
# ✅ merge_asof()에서 by의 역할
# 📌 요점 정리:
# on=: 병합 기준이 되는 정렬된 키 컬럼 (주로 시간/수량/가격 등 연속형)
# by=: 병합을 할 범주형(카테고리) 그룹 필터 — 같은 그룹끼리만 병합함

# 🎯 따라서,
# by=는 단순히 "정확히 일치"라기보다
# 👉 "같은 그룹 안에서만 범위 병합을 허용한다"는 의미야.

# 🚫 만약 by가 없으면?
# 수량 기준으로 전체 데이터가 비교됨
# 사이즈, 종류가 달라도 무조건 가장 가까운 수량끼리 병합됨 → 논리적 오류 발생 가능


# ✅ 바르게 고친 설명 문장
# by=는 범주형 그룹 조건을 지정해서, 같은 그룹 안에서만 범위 병합을 하도록 제한하는 인자입니다.
# 결과적으로 **"정확히 일치하는 그룹끼리만 병합"**이라는 효과를 줍니다

df5 = pd.merge_asof(df3, df4, on='수량', by=['사이즈', '종류'])
df5
```

```python
# 결과를 index는 제외하고 엑셀파일로 저장한다
df5.to_excel('result_merge_asof.xlsx', index=False)
```

## 7. 데이터 프레임 결합하기 (concat)



> pandas concat

pd.concat(objs, axis=0, join='outer', keys=None)

---

**objs** (인수는 시리즈 혹은 데이터프레임의 배열)

결합할 시리즈나 데이터프레임들을 리스트로 지정하는 인자

---

**axis** (인수는 0 또는 1 / 기본값은 0)

결합할 축 방향을 지정하는 인자.

 axis=0일때는 행이 증가하는 방향으로 결합하고,위↕아래로 붙임 (행이 늘어남)

 axis=1일때는 열이 증가하는 방향으로 결합한다.좌↔우로 붙임 (열이 늘어남)

---

**join** (인수는 ‘inner’, ‘outer’ / 기본값은 ‘outer’)

결합방식을 지정하는 인자.

* inner는 columns가 교집합으로 결합한다. (axis=0일때는 index가 교집합)
* outer는 columns가 합집합으로 결합한다. (axis=0일때는 index가 교집합)

---

**keys** (인수는 배열 / 기본값은 None)

결합하는 각 데이터프레임에 레벨을 부여하는 인자

---

```python
import pandas as pd

data1 = [[80, 69, 83, 98], [71, 90, 69, 66], [74, 72, 72, 95]]
data2 = [[68, 70, 84, 70], [65, 91, 90, 66], [78, 94, 96, 64]]
data3 = [[84, 70, 70, 68], [90, 66, 91, 65], [96, 64, 94, 78]]
idx1 = ['송중기', '박보검', '김나현']
idx2 = ['권보아', '김범수', '박효신']
col = ['국어', '수학', '영어', '과학']

df1 = pd.DataFrame(data1, index=idx1, columns=col)
df2 = pd.DataFrame(data2, index=idx2, columns=col)
df3 = pd.DataFrame([[65, 82], [85, 60], [75, 78]],
                   index=idx1, columns=['사회', '음악'])
df4 = pd.DataFrame(data3, index=idx2,
                   columns=['영어', '과학', '수학', '사회'])
```

```python
# 세로로 결합. 위↕아래로 붙임 (행이 늘어남)
pd.concat([df1, df2])
```

```python
# 가로로 결합. 좌↔우로 붙임 (열이 늘어남)
pd.concat([df1, df3], axis=1)
```

```python
# outer join으로 결합
pd.concat([df1, df4])
```

```python
# inner join으로 결합
pd.concat([df1, df4], join='inner')
```

```python
# 엑셀파일에서 3개의 시트를 병합하기
url = '/content/02_07_concat.xlsx'
df1 = pd.read_excel(url)
df2 = pd.read_excel(url, sheet_name=1)
df3 = pd.read_excel(url, sheet_name=2)
pd.concat([df1, df2, df3])
```

```python
# 딕셔너리에서 밸류(value)만 리스트에 담는 방법
dict_1 = {'key1': 'value1', 'key2':'value2'}
dict_1.values()
```

```python
# 엑셀파일의 모든 시트를 병합하는 코드
#강사왈:
# pd.read_excel(url, sheet_name=None)
#-> read_excel()의 sheet_name=None은
# 엑셀의 모든 시트를 불러서 시트 이름과 데이터 프레임을
# 키와 벨류로 묶은 딕셔너리로 반환한다.

#강사왈:
# pd.read_excel(url, sheet_name=None).values()
# ->  .values() 는 파이썬 함수다. 딕셔너리에 적용해 딕셔너리의 벨류(value)만 모아서
# 배열로 만들어 준다.
# 이렇게 배열로 만들어지면, 이 결과 자체가 데이터 프레임으로만 이뤄진 배열이라 concat에 입력될 수 있다.
# 즉, objs의 인수가 될 수 있따


pd.concat(pd.read_excel(url, sheet_name=None).values())
# 강사: 이 코드로 시트가 몇개이든 상관없이 전부 하나로 합쳐진다.
```
